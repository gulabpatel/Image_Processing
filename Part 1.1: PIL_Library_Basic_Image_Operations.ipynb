#7. Transposing

```
transpose_image1 = image1.transpose(Image.FLIP_LEFT_RIGHT)
transpose_image2 = image1.transpose(Image.FLIP_TOP_BOTTOM)
transpose_image3 = image1.transpose(Image.ROTATE_90)
transpose_image4 = image1.transpose(Image.ROTATE_180)
transpose_image5 = image1.transpose(Image.ROTATE_270)
transpose_image6 = image1.transpose(Image.TRANSPOSE)
plt.figure(figsize=(10,10))
plt.subplot(3,2,1)
plt.imshow(transpose_image1)
plt.title("FLIP_LEFT_RIGHT")
plt.subplot(3,2,2)
plt.imshow(transpose_image2)
plt.title("FLIP_TOP_BOTTOM")
plt.subplot(3,2,3)
plt.imshow(transpose_image3)
plt.title("ROTATE_90")
plt.subplot(3,2,4)
plt.imshow(transpose_image4)
plt.title("ROTATAE_180")
plt.subplot(3,2,5)
plt.imshow(transpose_image5)
plt.title("ROTATE_270")
plt.subplot(3,2,6)
plt.imshow(transpose_image6)
plt.title("TRANSPOSE")
```

#8. Resizing

There are some methods to do resizing

* BILINEAR
* NEAREST
* BOX
* HAMMING
* BICUBIC
* LANCZOS

```
newsize = (300,300)
plt.figure(figsize=(10,10))
resized_image1 = image1.resize(newsize, Image.BILINEAR)
resized_image2 = image1.resize(newsize, Image.NEAREST)
plt.subplot(1,2,1)
plt.imshow(resized_image1)
plt.title("BILINEAR")
plt.subplot(1,2,2)
plt.imshow(resized_image2)
plt.title("NEAREST")
```

#15. Image Enhancement

* Color Enhancement
* Sharpness Enhancement
* Contrast Enhancement
* Brightness Enhancement

```
from PIL import ImageEnhance
plt.figure(figsize=(10,10))
image_color_enhanc = image1.copy()
enhance_image1 = ImageEnhance.Color(image_color_enhanc).enhance(2.5)
enhance_image2 = ImageEnhance.Contrast(image_color_enhanc).enhance(2.5)
enhance_image3 = ImageEnhance.Brightness(image_color_enhanc).enhance(2.5)
enhance_image4 = ImageEnhance.Sharpness(image_color_enhanc).enhance(2.5)
plt.subplot(2,2,1)
plt.imshow(enhance_image1)
plt.title("Color")
plt.subplot(2,2,2)
plt.imshow(enhance_image2)
plt.title("Contrast")
plt.subplot(2,2,3)
plt.imshow(enhance_image3)
plt.title("Brightness")
plt.subplot(2,2,4)
plt.imshow(enhance_image4)
plt.title("Sharpness")
```

```
from PIL import ImageEnhance
plt.figure(figsize=(10,10))
image_color_enhanc = image1.copy()
enhance_image1 = ImageEnhance.Color(image_color_enhanc).enhance(.5)
enhance_image2 = ImageEnhance.Contrast(image_color_enhanc).enhance(.5)
enhance_image3 = ImageEnhance.Brightness(image_color_enhanc).enhance(.5)
enhance_image4 = ImageEnhance.Sharpness(image_color_enhanc).enhance(.5)
plt.subplot(2,2,1)
plt.imshow(enhance_image1)
plt.title("Color")
plt.subplot(2,2,2)
plt.imshow(enhance_image2)
plt.title("Contrast")
plt.subplot(2,2,3)
plt.imshow(enhance_image3)
plt.title("Brightness")
plt.subplot(2,2,4)
plt.imshow(enhance_image4)
plt.title("Sharpness")
```

#16. Alpha Blending

out = image1*(1.0-alpha) + image2*alpha

```
image1 = image1.copy()
image2 = image2.copy()

#Making both images same size
image2 = image2.resize(image1.size)

#Blending
image_blend = Image.blend(image1, image2, 0.3)
plt.imshow(image_blend)
```

```
image_blend = Image.blend(image1, image2, 0.8)
plt.imshow(image_blend)
```

#17. Image Transform

* 1. Affine Transform
* 2. Extent Transform
* 3. QUAD
* 4. MESH

```
image_transform = image1.copy()
image_transform = image_transform.transform(image_transform.size, Image.AFFINE, (1,-0.5,0.5*image_transform.size[0],0,1,0))
plt.imshow(image_transform)
```

```
image_transform = image1.copy()
image_transform = image_transform.transform(image_transform.size, Image.EXTENT, (100,100,image_transform.size[0],image_transform.size[1]//2))
plt.imshow(image_transform)
```

#18. Flipping Channels
