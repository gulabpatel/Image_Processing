```
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```
import numpy as np
```

```
img = cv.imread('image.jpg')
cv2_imshow(img)
```

##Convert into gray

```
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

#Laplacian

```
lap = cv.Laplacian(gray, cv.CV_64F)
lap = np.uint8(np.absolute(lap)) #pixel value can not be negative
cv2_imshow(lap)
```

#Sobel

##Sobelx

```
sobelx = cv.Sobel(gray, cv.CV_64F, 1 ,0)
cv2_imshow(sobelx)
```

Sobely

```
sobely = cv.Sobel(gray, cv.CV_64F, 0 ,1)
cv2_imshow(sobely)
```

##combined sobel

```
combined_sobel = cv.bitwise_and(sobelx,sobely)
cv2_imshow(combined_sobel)
```

#Canny

```
canny = cv.Canny(gray, 150, 175)
cv2_imshow(canny)
```
